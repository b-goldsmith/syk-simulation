# Generate Resource Estimates for Asymmetric Qubitization and QSP

This notebook can be used to generate resource estimates and store them in the `re_data` directory as a CSV file.

```python
import sys
import os

sys.path.insert(0, os.path.abspath("../.."))
```

```python
# This cell is being used to generate Asymmetric Qubitization and QSP resource estimates and writing them to the re_data folder. 
import numpy as np
from syk_simulation.qubitization.re_utils import gather_aq_estimates, gather_aq_re_cpp
import os

time = 1 
J = 1
epsilon = 5e-5
random_seed=3
extra_qubits = 28

filename = "re_data/scale_N_J1_r1.csv"

Nlist = [8, 12]

# Nlist = [8, 12, 16, 20, 24, 28, 32, 36, 48, 56, 64, 68, 80, 96, 100]

for N in Nlist:
    estimates = gather_aq_estimates(N=N, time=time, J=J, epsilon=epsilon, random_seed=3, extra_qubits=extra_qubits)
    # estimates = gather_aq_re_cpp(N=N, time=time, J=J, epsilon=epsilon, random_seed=3, extra_qubits=extra_qubits)

    rotation_eps = 1e-3
    total_t_gates = estimates['t_gates']+estimates['rotations']*(0.53* np.log2(1/rotation_eps) + 4.86)
                                            #  group['rotations']*(0.53* np.log2(1/(1 - (1-(group['epsilon']))**(1/ (group['rotations'])))) + 4.86))
    row = f"AQ,{N},{J},{time},{epsilon},{random_seed},{estimates['rotations']},{estimates['t_gates']},{total_t_gates},{estimates['qubit_highwater']}"

    with open(filename, "a", newline="") as f:
        f.write(f"{row}\n")
```

```python
# This cell is being used to generate Asymmetric Qubitization resource estimates and writing them to the re_data folder. 
import numpy as np
from syk_simulation.qubitization.re_utils import gather_aq_estimates, gather_aq_re_cpp
import csv
import os

time = 1 
J = 1
epsilon = 1e-3
random_seed=3
extra_qubits = 25

filename = "re_data/scale_eps_J1_order1.csv"

N = 64 

Epsilonlist = [5e-4, 5e-5, 5e-6, 5e-7, 5e-8, 5e-9, 5e-10, 5e-11, 5e-12, 5e-13]

for epsilon in Epsilonlist:
    estimates = gather_aq_estimates(N=N, time=time, J=J, epsilon=epsilon, random_seed=3, extra_qubits=extra_qubits)
    # estimates = gather_aq_re_cpp(N=N, time=time, J=J, epsilon=epsilon, random_seed=3, extra_qubits=extra_qubits)

    rotation_eps = epsilon
    total_t_gates = estimates['t_gates']+estimates['rotations']*(0.53* np.log2(1/rotation_eps) + 4.86)
    row = f"AQ,{N},{J},{time},{epsilon},{random_seed},{estimates['rotations']},{estimates['t_gates']},{total_t_gates},{estimates['qubit_highwater']}"

    with open(filename, "a", newline="") as f:
        f.write(f"{row}\n")
```
