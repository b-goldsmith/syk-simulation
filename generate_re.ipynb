```python
# This cell is being used to generate Asymmetric Qubitization resource estimates and writing them to the re_data folder. 
import numpy as np
from syk_simulation.qubitization.re_utils import gather_aq_estimates
import csv

time = 1 
J = 1
epsilon = 1e-3
random_seed=3

extra_qubits = 25


Nlist = [8, 12, 16, 20, 24, 28, 32, 48, 56, 64, 68, 80, 96, 112, 128, 132]

for N in Nlist:
    estimates = gather_aq_estimates(N=N, time=time, J=J, epsilon=epsilon, random_seed=3, extra_qubits=extra_qubits)

    rotation_eps = 1e-3
    total_t_gates = estimates['t_gates']+estimates['rotations']*(0.53* np.log2(1/rotation_eps) + 4.86)
    row = f"AQ,{N},{J},{time},{epsilon},{random_seed},{estimates['rotations']},{estimates['t_gates']},{total_t_gates},{estimates['qubit_highwater']}"

    filename = "syk_simulation/notebooks/re_data/scale_N_J1_order1.csv"
    with open(filename, "a", newline="") as f:
        writer = csv.writer(f)
        writer.writerow([row])
```
